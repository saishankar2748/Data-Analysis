### II) Superstore Data

In the `superstore.csv` file, you have the details of orders purchased in an American online retail store. Load the dataset, observe and analyse the different columns and answer the following questions.

```python
#Load the dataset
df2 = pd.read_csv('superstore.csv')
df2
```

```python
df2[df2['Profit']==df2['Profit'].max()]
```

```python
df2.groupby('Product ID')['Profit'].max().sort_values()
```

```python
df2[df2['Product ID']=='TEC-CO-10004722']
```

```python
sns.jointplot(df2['Sales'],df2['Profit'],kind='reg')
```

```python
df2[(df2['Sales']<15000) & (df2.Profit<0)].plot(kind='scatter',x='Sales',y='Profit')
```

```python
df3 = df2[(df2.Profit < 0) & (df2.Sales < 15000)]
sns.jointplot('Sales', 'Profit', df3)
plt.show()
```

```python
df2.groupby(['Segment','Ship Mode'])[['Profit']].mean()
```

```python
res = pd.pivot_table(data=df2,index='Segment',columns='Ship Mode',values='Profit')
res
```

```python
res = pd.pivot_table(data=df2,index='Segment',columns='Ship Mode',values='Profit')
sns.heatmap(res,annot=True)
```

```python
sns.barplot(data=df2,x='Segment',y='Sales')
```
